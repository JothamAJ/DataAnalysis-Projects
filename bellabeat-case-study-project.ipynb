<div style="background-color: #0000FF; padding: 10px; color: #FFFFFF; font-size: 16px; font-weight: bold;">USERS SLEEP EFFECIENCY</div>

- Sleep effeciency is the ratio of the total time asleep compared to the total time one spent in bed
- Analysing sleep effeciency may lead to insights into the behavioural pattens of the users. Low sleep efficiency is sometimes linked with poor overall performance

```python
#Calculate and analyse sleep efficiency 
sleep_day_df['SleepEfficiency'] = round((sleep_day_df['TotalMinutesAsleep'] / sleep_day_df['TotalTimeInBed']) * 100)
sleep_day_df.head()
```

```python
bins = [0, 80, 85, 100]
labels = ['Poor', 'Fair', 'Good']

sleep_day_df['SleepCategory'] = pd.cut(sleep_day_df['SleepEfficiency'], bins=bins, labels=labels, include_lowest=True)

#Count the number of occurrences in each category
counts = sleep_day_df['SleepCategory'].value_counts()

colors = ['#d3ffce','#fa8072', '#c0c0c0']

explode = [0, 0, 0.1]

plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90, colors=colors, explode=explode)
plt.title('Sleep Efficiency')
plt.show()
```

<div style="background-color: #0000FF; color: #FFFFFF; font-size: 14px; font-weight: bold;">FINDINGS FROM PIE CHART</div>
The Pie chart has helped us: 

1. Over 90% of the users have good sleep efficiency
2. Sleep efficieny will be or further analysis
3. Relation between sleep efficiency and activity levels? weight?

<div style="background-color: #0000FF; padding: 10px; color: #FFFFFF; font-size: 16px; font-weight: bold;">USERS DAILY ACTIVITY PATTERNS</div>

- Categorise the average total steps to find possible trends
- See if they link with users sleep efficiency and maybe BMI

```python
#Analyse daily activity patterns
#Explore activity intensity distribution
#Categorise users activity levels based on certain conditions: 
#Average steps in UK is 4,000
#sedetary: <4000 steps on average
#active: 4000 - 10,000 steps on average
#very active: > 10,000 steps on average

# Conditiion values by number of steps
mean_steps_df = daily_activity_df.groupby('Id')['TotalSteps'].mean().reset_index()
mean_steps_df = mean_steps_df.round()
conditions = [
    (mean_steps_df['TotalSteps'] <= 4000),
    (mean_steps_df['TotalSteps'] > 4000) & (mean_steps_df['TotalSteps'] < 10000),
    (mean_steps_df['TotalSteps'] >= 10000)
]

values = ['sedentary', 'active', 'very active']
mean_steps_df['ActivityLevel'] = np.select(conditions, values)
mean_steps_df.head(5)
```

```python
#Add to main data set
daily_activity_level_df = pd.merge(daily_activity_df, mean_steps_df[['Id', 'ActivityLevel']], on='Id', how='left')
daily_activity_level_df.head(5)
```

```python
#steps per day of week
day_activity_df = daily_activity_level_df.groupby(['DayOfWeek', 'ActivityLevel'])['TotalSteps'].mean().reset_index()
days_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

palette = sns.color_palette("viridis", len(day_activity_df['ActivityLevel'].unique()))

#plotting
plt.figure(figsize=(12, 8))
sns.barplot(x='DayOfWeek', y='TotalSteps', hue='ActivityLevel', order=days_order, data=day_activity_df, palette=palette)

#labels
plt.title('Average Steps by Activity Level per Day of the Week', fontsize=16)
plt.xlabel('Day of the Week', fontsize=14)
plt.ylabel('Average Steps', fontsize=14)

# Legend
plt.legend(title='Activity Level', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=12)

# Grid lines
plt.grid(axis='y', linestyle='--', alpha=0.7)

#average steps line
overall_mean = day_activity_df['TotalSteps'].mean()
plt.axhline(overall_mean, color='red', linestyle='--', linewidth=2, label='Overall Mean')


# Show the plot
plt.show()
```

Findings:

```python
#BMI Categories
bins = [0, 18.5, 25, 30, 35, 40, float('inf')]
labels = ['Underweight', 'Normal', 'Overweight', 'Obese Class I', 'Obese Class II', 'Severe Obesity']

#add column
weight_activity_df['BMICategory'] = pd.cut(weight_activity_df['BMI'], bins=bins, labels=labels, include_lowest=True)
avr_steps_per_bmi = weight_activity_df.groupby('BMICategory')['TotalSteps'].mean().reset_index()

#plot
plt.figure(figsize=(12, 8))
sns.barplot(x='BMICategory', y='TotalSteps', data=avr_steps_per_bmi, palette=colour, saturation=0.7)

#labels
plt.title('Average Total Steps per BMI Category', fontsize=16)
plt.xlabel('BMI Category', fontsize=14)
plt.ylabel('Average Total Steps', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)

for index, value in enumerate(avr_steps_per_bmi['TotalSteps']):
    plt.text(index, value + 50, f'{value:.2f}', ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

2 -  Users are mostly active on Tuesdays and less so on Sundays. FITBIT should take advantage of this and schedule their promotions towards this new information.

3 - Those with better sleep efficency average less steps, they may priortise their sleep over exercising. FITBIT should take this into account and provide programs that educate and promote the importance of slee. Workout schedules should also be flexible to suit these users. FITBIT should provide users with a more tailor made and personalied fitness plan to fit their routine.

4 - The tracking features were unreliable and incosistent. FITBIT should pay more attention to this so users consistently report any changes in weight.

5 - FITBIT should providemore detailed and descriptive programs to suit users that may be deemed as obese. Provide ways for them to stay engaged throughout the program. Clear outcomes, goals and rewards should motivate these users so that in the future they are able to imporve their step count.
